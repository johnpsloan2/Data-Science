# What's the most expensive listing? What else can you tell me about the listing?

The most expensive listing was a house listed for $1,000.00 in Ravenna (Listing ID 4825073). The listing was for the entire home that accommodates 4 guests. The home has 2 bedrooms with 2 real beds and 1 bathroom. This home has only been listed once on the AirBnB market, and the host received perfect reviews by the guests of that stay. 

Interstingly enough, this home went for $25 cheaper than a home that accommodates 11 guests with 5 bedrooms, 7 beds (all real beds), and 4.5 baths. This home was located in Queen Anne. 

SELECT 
	
    id,
	
    price,
	
    neighbourhood,
	
    property_type,
	
    room_type,
	
    accommodates,
	
    bathrooms,
	
    bedrooms,
	
    beds,
	
    bed_type,
	
    review_scores_rating,
	
    host_total_listings_count

FROM 
	
    listings

ORDER by price DESC

LIMIT 3

# What neighborhoods seem to be the most popular?

Based on number of listings, the top 5 neighbourhoods would be Capitol Hill (351), Ballard (213), Belltown 204, Minor (192), and Queen Anne (187). 


SELECT 
	
    neighbourhood,
	
    count(*) 'Number of Listings'

FROM
	
    listings

WHERE neighbourhood != ''

GROUP by 1

ORDER by 2 desc


# What time of year is the cheapest time to go to your city? What about the busiest?

The cheapest time to go to Seattle, Washington is in January. The average price of listing in January. It comes in at an average listing price of $69 which is $13 cheaper than the next cheapest month, Feburary.

SELECT 
	
    avg(price) 'Average Monthly Price',
	
    strftime('%m', date) as Month

FROM
	
    calendar

GROUP by 2

ORDER by 1 DESC

March, May, July, August, and December are each equally busy according to the data. All of these months come in at a listings total for the month of 118,358; however, December seems to have the most competition resulting higher prices compared to the other months.

SELECT 
	
    strftime('%m', date) as Month,
	
    count(*)

FROM
	
    calendar

GROUP by 1

ORDER by 2 DESC
