# Joins and CTEs


1 - What are the three longest trips on rainy days?

WITH 
	
    Rdays 

AS 

(	
	
    SELECT 
		
        date
	
    FROM 
		
        weather
	
    WHERE 
		
        events = 'Rain'
	
    GROUP BY 1

)

SELECT
	
    trip_id,
	
    duration,

DATE
	
    (start_date) trip_date

FROM 
	
    trips

JOIN 
	
    rdays

ON
	
    rdays.date = trip_date

ORDER BY 
	
    duration DESC



2 - Which station is empty most often?


SELECT 
    
    status.station_id,
    
    stations.name, 

COUNT(CASE WHEN docks_available=0 then 1 END) empty_count 

FROM 
    
    status 

JOIN 
    
    stations on stations.station_id = status.station_id 

GROUP BY 1 

ORDER BY empty_count DESC






3 - Return a list of stations with a count of number of trips starting at that station but ordered by dock count.


SELECT 
	
    trips.start_station,
	
    count(*) station_count,
	
    stations.dockcount

FROM 
	
    trips

JOIN 
	
    stations

ON 
	
    trips.start_station = stations.name

GROUP by trips.start_station

ORDER by stations.dockcount



4 - What's the length of the longest trip for each day it rains anywhere?




WITH 
    
    rdays 

AS 

(

SELECT 
    
    date

FROM 
    
    weather

WHERE 
    
    events = 'Rain'

GROUP BY 1

),

rday_trips AS 

(

SELECT
    
    trip_id,
    
    duration,

DATE(start_date) trip_date

FROM trips

JOIN rdays

ON rdays.date = trip_date

ORDER BY duration DESC
)

SELECT 

trip_date,

max(duration)

from rday_trips

GROUP BY 1
