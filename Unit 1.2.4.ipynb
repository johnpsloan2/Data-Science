# Joins and CTEs
WITH 
	Rdays 
AS 
(	
	SELECT 
		date
	FROM 
		weather
	WHERE 
		events = 'Rain'
	GROUP BY 1

)

SELECT
	trip_id,
	duration,
DATE
	(start_date) trip_date
FROM 
	trips
JOIN 
	rdays
ON
	rdays.date = trip_date
ORDER BY 
	duration DESC
LIMIT 3






SELECT 
	stations.name,
	count(*) as empty_count
FROM 
	stations
JOIN 
	status
ON
	stations.station_id = status.station_id
WHERE 
	status.bikes_available = 0
GROUP by stations.name
ORDER by empty_count DESC









SELECT 
	trips.start_station,
	count(*) station_count,
	stations.dockcount
FROM 
	trips
JOIN 
	stations
ON 
	trips.start_station = stations.name
GROUP by trips.start_station
ORDER by stations.dockcount

rday_trips as (
