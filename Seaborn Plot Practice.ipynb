```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
%matplotlib inline
```

```python
sales = pd.read_csv('vgsales.csv')
```

```python
top_sales = sales.drop(sales.index[101:])
```

# Choose one variable and plot that variable four different ways.
First, I choose a histogram.

```python
histogram = sns.distplot(top_sales['NA_Sales'], bins=15)
```

Next, just a density distribution plot.

```python
desnity = sns.distplot(top_sales['NA_Sales'], hist=False, rug=True)
```

Third, a box plot.

```python
sns.set_style("whitegrid")
box_plot = sns.boxplot(x=top_sales['NA_Sales'])
```

Lastly, a histogram with a rugplot 

```python
sns.distplot(top_sales['NA_Sales'], kde=False, rug=True, bins=15)
```

# Choose two continuous variables, and plot them three different ways.
First, a scatter plot.

```python
scatter = sns.jointplot(y='NA_Sales',
               x='EU_Sales',
               data=top_sales)
```

Second, a scatter plot with regression line. 

```python
reg_scatter = scatter = sns.jointplot(y='NA_Sales',
               x='EU_Sales',
               data=top_sales, kind='reg')
```

Next, a density scatter plot.

```python
Density_scatter = (sns.jointplot(y='NA_Sales',
               x='EU_Sales',
               data=top_sales, color='g')
            .plot_joint(sns.kdeplot, zorder=0, n_levels=6))
```

Last, a density graph without scatter. 

```python
Density = sns.jointplot('NA_Sales', 'EU_Sales', data=top_sales,
                       kind='kde', color='purple')
```

# Choose one continuous variable and one categorical variable, and plot them six different ways.

First, a strip plot.

```python
fig_dims = (20,5)
fig, ax = pyplot.subplots(figsize=fig_dims)
sns.stripplot(x='Genre', y='Global_Sales', data=top_sales) 
```

Second, a bar chart.

```python
fig_dims = (20,5)
fig, ax = pyplot.subplots(figsize=fig_dims)
sns.barplot(x='Genre', y='Global_Sales', data=top_sales)
```

Third, a swarm plot.

```python
fig_dims = (20,5)
fig, ax = pyplot.subplots(figsize=fig_dims)
sns.swarmplot(x="Genre", y="Global_Sales", data=top_sales)
```

Next, a violin plot.

```python
fig_dims = (20,5)
fig, ax = pyplot.subplots(figsize=fig_dims)
sns.violinplot(x="Genre", y="Global_Sales", data=top_sales)
```

Fifth, a box plot.

```python
fig_dims = (20,15)
fig, ax = pyplot.subplots(figsize=fig_dims)
sns.boxplot(x="Genre", y="Global_Sales", data=top_sales)
```

Sixth, a strip plot with jitter. 

```python
fig_dims = (20,5)
fig, ax = pyplot.subplots(figsize=fig_dims)
sns.stripplot(x="Genre", y="Global_Sales", data=top_sales, jitter=True)
```
